```python
from citipy import citipy
import numpy as np
import requests
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Incorporated citipy to determine city based on latitude and longitude
from config import api_key
```

```python
weather_url = "http://api.openweathermap.org/data/2.5/weather?units=imperial&"
cities = []
latitude = []
lontiude = []
wind =[]
cloud =[]
humidity=[]
temp=[]

while len(cities) <=500:
    lat = np.random.randint(-90, 90)
    lng = np.random.randint(-180,180)
    city = citipy.nearest_city(lat, lng).city_name
    if city not in cities:
        try:
            url = weather_url+ "appid=" + api_key + "&q="+city
            response = requests.get(url).json()
            latitude.append(response["coord"]["lat"])
            lontiude.append(response["coord"]["lon"])
            wind.append(response["wind"]["speed"])
            cloud.append(response["clouds"]["all"])
            humidity.append(response["main"]["humidity"])
            temp.append(response["main"]["temp"])
            cities.append(city)
            print(f"Proceed {len(cities)} : {city} ")
        except:
            pass

    
    
    
```

```python
df = pd.DataFrame({"city":cities
                  ,"latitude":latitude
                  ,"lontitude":lontiude
                  ,"wind speed":wind
                  ,"cloudness":cloud
                  ,"humidity":humidity
                  ,"temp":temp})
df
```

```python
# * Temperature (F) vs. Latitude
df.plot.scatter( x="latitude",y= "temp")
plt.title("Temperature (F) vs. Latitude(2/2/20)")
plt.xlabel("Latitude")
plt.ylabel("Temperature (F)")
# plot scatter polt using Temperature and Latitude, it looks like they have Curvilinear correlation.
plt.savefig("temp_Latitude.png")
```

```python
# * Humidity (%) vs. Latitude
df.plot.scatter( x="latitude",y= "humidity")
plt.title("humidity (%) vs. Latitude(2/2/20)")
plt.xlabel("Latitude")
plt.ylabel("humidity (%) ")
# plot scatter polt using humidity and Latitude, it looks like they don't have correlation.
plt.savefig("humidity_Latitude.png")
```

```python
# * Cloudiness (%) vs. Latitude
df.plot.scatter( x="latitude",y= "cloudness")
plt.title("Cloudiness (%) vs. Latitude(2/2/20)")
plt.xlabel("Latitude")
plt.ylabel("Cloudiness (%) ")
# plot scatter polt using Cloudiness and Latitude, it looks like they don't have correlation.
plt.savefig("cloudness_Latitude.png")
```

```python
# * Wind Speed (mph) vs. Latitude
df.plot.scatter( x="latitude",y= "wind speed")
plt.title("wind speed (mph) vs. Latitude(2/2/20)")
plt.xlabel("Latitude")
plt.ylabel("Wind Speed (mph) ")
# plot scatter polt using wind speed and Latitude, it looks like they don't have correlation.
plt.savefig("wind_speed_Latitude.png")
```

```python
# * Southern Hemisphere - Humidity (%) vs. Latitude
# * Northern Hemisphere - Cloudiness (%) vs. Latitude
# * Southern Hemisphere - Cloudiness (%) vs. Latitude
# * Northern Hemisphere - Wind Speed (mph) vs. Latitude
# * Southern Hemisphere - Wind Speed (mph) vs. Latitude

north_df = df.loc[df["latitude"]>=0]
south_df = df.loc[df["latitude"]<0]

```

```python
# * Northern Hemisphere - Temperature (F) vs. Latitude
x_values = north_df['latitude']
y_values = north_df['temp']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(min(x_values),min(y_values)),fontsize=15,color="red")
plt.xlabel("Latitude")
plt.ylabel("Temperature (F)")
print(f"The r-squared is: {rvalue}")
plt.savefig("lin_temp_NLatitude.png")
plt.show()
```

```python
# * Southern Hemisphere - Temperature (F) vs. Latitude
x_values = south_df['latitude']
y_values = south_df['temp']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(min(x_values),min(y_values)),fontsize=15,color="red")
plt.xlabel("Latitude")
plt.ylabel("Temperature (F)")
print(f"The r-squared is: {rvalue}")
plt.savefig("lin_temp_SLatitude.png")
plt.show()
# based on r-squared value and lin , Northern Hemisphere and Temperature (F) has positive correlation
# based on r-squared value and lin , Southern Hemisphere and Temperature (F) has negative correlation
```
